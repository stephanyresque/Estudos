## Merge

```python
PATH = r"C:\Users\steph\OneDrive\Documentos\MeusProjetos\Estudos\pandas\data\customers.csv"

df_customer = pd.read_csv(PATH, sep = ';')
```

```python
PATH = r"C:\Users\steph\OneDrive\Documentos\MeusProjetos\Estudos\pandas\data\transactions.xlsx"

df_transaction = pd.read_excel(PATH)
```

```python
df_transaction.merge(df_customer,
                    how = 'inner',
                    left_on="IdCustomer",
                    right_on="UUID",
                    suffixes=["_transacao", "_cliente"])
```

```python
PATH = r"C:\Users\steph\OneDrive\Documentos\MeusProjetos\Estudos\pandas\data\transactions_cart.parquet"

df_transaction_product =  pd.read_parquet(PATH)
```

```python
df_join = df_transaction.merge(df_customer,
                    how = 'inner',
                    left_on="IdCustomer",
                    right_on="UUID",
                    suffixes=["_transacao", "_cliente"])
```

```python
df_join.merge(df_transaction_product,
              how = 'inner',
              left_on='UUID_transacao',
              right_on='IdTransaction')
```

## Concat

```python
data_01 = {
    "Id":[1,2,3,4,5],
    'Nome':['Teo', 'Maria', 'Lara', 'Nath', 'Ste'],
    'Idade':[32,22,56,21,21],
}

df_01 = pd.DataFrame(data_01)

data_02 = {
    "Id":[6,7,8,9,10],
    'Nome':['João', 'Nathan', 'Lucas', 'Mario', 'Fernando'],
    'Idade':[23,25,65,50,18],
}

df_02 = pd.DataFrame(data_02)
```

```python
pd.concat( [df_01, df_02] ).reset_index(drop=True)
```

```python
data_03 = {
    'Sobrenome':['Costa', 'Silva', 'Souza', 'Moreno', 'Resque'],
    'Renda':[3200,2200,5600,2100,2123],
}

df_03 = pd.DataFrame(data_03)
```

```python
df_03
```

```python
df_03 = df_03.sort_values(['Renda', 'Sobrenome'], ascending=[False, True])
```

```python
pd.concat([df_01,df_03], axis = 1)
```
