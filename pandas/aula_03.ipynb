```python
PATH = r"C:\Users\steph\OneDrive\Documentos\MeusProjetos\Estudos\pandas\data\customers.csv"

df = pd.read_csv(PATH, sep =';')
```

```python
df.head()
```

```python
df['Points_double'] = df['Points'] * 2
df
```

```python
df['Points_ratio'] = df['Points_double'] / df['Points']
df
```

```python
df['Const'] = 1
df
```

```python
df['Nome_alto'] = df['Name'].str.upper()
df
```

```python
def get_first(nome):
    return nome.split("_")[0]

get_first("ste_resque")
```

```python
df['primeiro_nome'] = df['Name'].apply(get_first)
df.head()
```

```python
df['ultimo_nome'] = df['Name'].apply(lambda x:x.split("_")[-1])
df.head()
```
