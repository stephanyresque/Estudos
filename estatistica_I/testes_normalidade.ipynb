```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
dados = norm.rvs(size = 1000)
dados
```

```python
plt.hist(dados, bins = 20)
plt.title("Dados")

plt.show()
```

```python
fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show()
```

```python
stats.shapiro(dados)
```

```python
dados_2 = skewnorm.rvs(4, size = 1000)
```

```python
plt.hist(dados_2, bins = 20)
plt.title("Dados")
plt.show()
```

```python
fig, ax = plt.subplots()
stats.probplot(dados_2, fit = True, plot = ax)
plt.show()
```

```python
stats.shapiro(dados_2)
```
