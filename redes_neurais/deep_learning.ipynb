```python
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
import numpy as np
from sklearn.metrics import confusion_matrix
from keras.datasets import mnist
```

```python
(X_train, X_test), (y_train, y_test) = mnist.load_data()

plt.imshow(X_train[21], cmap = 'gray')
plt.title(y_train[21])
```

```python
# Mudança de dimensão, originalmente está em 28x28 e precisamos 784
X_train = X_train.reshape((len(X_train), int(np.prod(X_train.shape[1:]))))
X_test = X_test.reshape((len(X_test), int(np.prod(X_test.shape[1:]))))
X_test[0]
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
# Normalização (255 é o valor máximo de um pixel)
X_train /= 255
X_test /= 255
```

```python
# Transformação para o formato dummy (temos 10 classes)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
y_test[0]
```
